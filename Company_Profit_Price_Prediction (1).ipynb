#DATA PREPROCESSING

```python
# data Preprocessing
```

```python
df.isnull().sum()
# finding whether the data frames containing null values
```

```python
df = df.dropna()
# dropping all the null values
```

```python
# dropping the duplicate value present in the dataframe
df.drop_duplicates()
```

```python
df = df.replace(',',"",regex=True)
# replacing ',' using regular expressions
```

```python
df
```

```python
x = df.iloc[:,[1,2,3,4,5,6,7]]
y = df.iloc[:,8]
# assigning x and y values
```

```python
df.dtypes
# give sthe data types
```

```python
df.describe()
# give the statistical information of the data
```

# MODEL MAKING WITH DECISION TREE

```python
# making model with decision tree
```

```python
from sklearn import tree
model = tree.DecisionTreeRegressor()
```

```python
# As our dataset contains very less number of rows we canot split the data
```

```python
mymodel = model.fit(x,y)
# Training the model
```

```python
pred_y = mymodel.predict(x)
# predicting the y values
```

```python
pred_y
```

# Model Evaluation

```python
from sklearn.metrics import r2_score
r2_score(pred_y,y)
# give  the perfomace of the model
```
